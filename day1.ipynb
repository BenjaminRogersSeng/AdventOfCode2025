```python
number = 50
zero_count = 0

with open('input.txt', 'r') as file:
        for line in file:
                line = line.strip()
                going_left = True if line.find('L') else False
                turns = int(line.strip('LR'))
                if going_left:
                 turns = -turns

                while number + turns < 0 or number + turns > 99:

                     if going_left and number + turns < 0:
                         number += 100
    
                     if not going_left and number + turns > 99:
                         number -= 100
 
                number += turns
                
                if number == 0:
                        zero_count += 1
print(zero_count)
```
